```python
# Import the pandas library for data manipulation
import pandas as pd

# Load the CSV file into a pandas DataFrame
student_performance_df = pd.read_csv('/content/Student_performance_10k.csv')

# Display the first few rows of the DataFrame
display(student_performance_df.head())
```

```python
# Get the number of rows and columns in the DataFrame
rows, columns = student_performance_df.shape

# Print the number of rows and columns
print(f"The DataFrame has {rows} rows and {columns} columns.")
```

```python
# Get a list of the column names
print(student_performance_df.columns.tolist())
```

```python
# Display data types of each column
display(student_performance_df.dtypes)
```

```python
# Check for null values in each column and sum them
print(student_performance_df.isnull().sum())
```

```python
# Drop rows with missing values
student_performance_df_cleaned = student_performance_df.dropna()

# Check if there are any missing values left
print("Missing values after dropping rows:")
print(student_performance_df_cleaned.isnull().sum())

# Display the shape of the cleaned DataFrame
print("\nShape of the DataFrame after dropping rows:")
print(student_performance_df_cleaned.shape)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize the distribution of numerical scores
numerical_cols = ['math_score', 'reading_score', 'writing_score', 'science_score', 'total_score']
for col in numerical_cols:
    plt.figure(figsize=(8, 5))
    sns.histplot(student_performance_df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

# Visualize the distribution of categorical variables
categorical_cols = ['gender', 'race_ethnicity', 'parental_level_of_education', 'lunch', 'test_preparation_course', 'grade']
for col in categorical_cols:
    plt.figure(figsize=(8, 5))
    sns.countplot(data=student_performance_df, x=col)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
# Drop the 'roll_no' column as it's not relevant for correlation analysis
student_performance_df_numeric = student_performance_df.drop('roll_no', axis=1)

# Calculate and plot the correlation matrix
sns.heatmap(student_performance_df_numeric.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
